```python
# Univariate analysis

# Histogram for numerical columns (e.g., Sales, Quantity, Discount, Profit)
plt.figure(figsize=(10, 6))
sns.histplot(df['Sales'], bins=30, kde=True)
plt.title('Distribution of Sales')
plt.xlabel('Sales')
plt.ylabel('Frequency')
plt.show()
```

```python
# Bar plot for categorical columns (e.g., Category, Sub-Category, Segment)
plt.figure(figsize=(10, 6))
sns.countplot(x='Category', data=df)
plt.title('Count of Products in Each Category')
plt.xlabel('Category')
plt.ylabel('Count')
plt.xticks(rotation=45)
```

```python
# Bivariate analysis

# Scatter plot to visualize the relationship between Sales and Profit
plt.figure(figsize=(10, 6))
sns.scatterplot(x='Sales', y='Profit', data=df)
plt.title('Sales vs. Profit')
plt.xlabel('Sales')
plt.ylabel('Profit')
```

```python
# Box plot to compare Profit across different Categories
plt.figure(figsize=(10, 6))
sns.boxplot(x='Category', y='Profit', data=df)
plt.title('Profit by Category')
plt.xlabel('Category')
plt.ylabel('Profit')
plt.xticks(rotation=45)
```

```python
# Multivariate analysis

# Pair plot to visualize relationships between numerical variables (e.g., Sales, Quantity, Discount, Profit)
sns.pairplot(df[['Sales', 'Quantity', 'Discount', 'Profit']])
```

```python
# Heatmap to visualize correlations between numerical variables
plt.figure(figsize=(10, 6))
sns.heatmap(df[['Sales', 'Quantity', 'Discount', 'Profit']].corr(), annot=True, cmap='coolwarm', fmt='.2f')
plt.title('Correlation Heatmap')
```

```python
# Filter rows with negative profit
negative_profit_df = df[df['Profit'] < 0]
```

```python
negative_profit_df
```

```python
# Group by sub-category and calculate the average profit and sales for each sub-category
subcat_profit_sales = negative_profit_df.groupby('Sub-Category')['Profit', 'Sales'].mean().reset_index()
```

```python
# Sort sub-categories by average profit in ascending order
subcat_profit_sales_sorted = subcat_profit_sales.sort_values(by='Profit', ascending=True)
```

```python
# Plot the sub-categories with negative profit and their average sales
plt.figure(figsize=(12, 6))
sns.barplot(x='Profit', y='Sub-Category', data=subcat_profit_sales_sorted, color='red')
plt.title('Sub-Categories with Negative Profit and their Average Sales')
plt.xlabel('Average Profit')
plt.ylabel('Sub-Category')
plt.show()
```

```python
df['Discount'].unique()
```

```python
#Cost Analysis
# Calculate the total cost of goods sold (COGS) for each sub-category with negative profit
subcat_negative_profit_cogs = negative_profit_df.groupby('Sub-Category')['Profit'].sum()
```

```python
#Pricing Analysis
# Calculate the average discount for each sub-category with negative profit
subcat_negative_profit_avg_discount = negative_profit_df.groupby('Sub-Category')['Discount'].mean()
```

```python
#Sales and Demand Analysis
# Calculate the average sales for each sub-category with negative profit
subcat_negative_profit_avg_sales = negative_profit_df.groupby('Sub-Category')['Sales'].mean()
```

```python
# Visualization
plt.figure(figsize=(12, 6))
sns.barplot(x='Sub-Category', y='Sales', data=negative_profit_df, estimator=lambda x: x.mean(), ci=None, color='blue', label='Average Sales')
sns.barplot(x='Sub-Category', y='Profit', data=negative_profit_df, estimator=lambda x: x.mean(), ci=None, color='red', label='Average Profit')
plt.title('Average Sales and Profit for Sub-Categories with Negative Profit')
plt.xlabel('Sub-Category')
plt.ylabel('Average Value')
plt.xticks(rotation=45)
plt.legend()
plt.show()
```

```python
#Analyze the impact of discount on Sales and Profit

#Scatter plot to visualize the relationship between Sales and Discount
plt.figure(figsize=(10, 6))
sns.scatterplot(x='Discount', y='Sales', data=df)
plt.title('Sales vs. Discount')
plt.xlabel('Discount')
plt.ylabel('Sales')
plt.show()
```

```python
#Compare discount levels for profitable and non-profitable transactions

# Create a new column to categorize transactions as profitable or non-profitable
df['Profitable'] = df['Profit'] >= 0
```

```python
# Box plot to compare discount levels for profitable and non-profitable transactions
plt.figure(figsize=(10, 6))
sns.boxplot(x='Profitable', y='Discount', data=df)
plt.title('Discount Distribution for Profitable and Non-Profitable Transactions')
plt.xlabel('Profitable')
plt.ylabel('Discount')
plt.xticks([0, 1], ['Non-Profitable', 'Profitable'])
plt.show()
```
